```python
import matplotlib.pyplot as plt

#to give the datapoints for them
x=[20, 15, 10]
y=[5, 10, 20]

#To plot  the point
plt.plot(x, y)

#Naming the axis
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#giving the name to the graph
plt.title('First Graph')

#function used to show the graph
plt.show()

#the graph is in an decrement way
```

```python
plt.plot(x, y, color="b", marker='o', linestyle='dashed', linewidth=2, markersize= 15)
```
